```python
from sympy import Symbol, integrate, cos, solve, Eq, sin, pi, diff, exp, sqrt
import scipy
import numpy as np
import matplotlib.pyplot as plt
```

**Упражнение 1**



```python
x = Symbol('x')
I = integrate(x*cos(x),(x, 0, pi/2))
print(f"S = {I} = {I.evalf():.3f}")

xv = np.linspace(0, np.pi/2, 1000)
fv = xv*np.cos(xv)
plt.plot(xv, fv)
plt.grid()
plt.show()
```

Упражнение 2




```python
x = Symbol('x')
fs1 = x**2+2*x
fs2 = 7-4*x-x**2
intersect = solve(Eq(fs1, fs2), x)
if len(intersect) == 0:
    print("Графики не пересекаются")
elif len(intersect) == 1:
    print(f"Графики пересекаются в одной точке: {intersect[0]}")
elif len(intersect) > 2:
    print(f"Графики пересекаются в большом количестве точек: {' '.join(intersect)}")
else:
    xv = np.linspace(intersect[0], intersect[1], 1000)
    fv1 = xv**2+2*xv
    fv2 = 7-4*xv-xv**2
    print(f"График пересекается в двух точках: {intersect[0]}, {intersect[1]}")
    integ = integrate(fs1-fs2, (x, intersect[0], intersect[1]))
    print(f"S = {integ} = {integ.evalf():.3f}")
    plt.plot(xv, fv1, "b")
    plt.plot(xv, fv2, "r")
    plt.axis("equal")
    plt.grid()
    plt.show()
    
```

**Упражнение 3**



```python
# А)
t = Symbol('t')
x = cos(t)
y = sin(t)
a = 0
b = 2*pi
integr = integrate(x*diff(y, t, 1), (t, a, b/2)) - integrate(-x*diff(y, t, 1), (t, b/2, b))
print(f"S = {integr} = {integr.evalf():.2f}")
tv = np.linspace(0, 2*np.pi, 1000)
xv = np.cos(tv)
yv = np.sin(tv)
plt.plot(xv, yv)
plt.axis("equal")
plt.grid()
plt.title("А) cos(t), sin(t)")
plt.show()

# Б)
t = Symbol('t')
x = 2*cos(t)
y = 3*sin(t)
a = 0
b = 2*pi
integr = integrate(x*diff(y, t, 1), (t, a, b/2)) - integrate(-x*diff(y, t, 1), (t, b/2, b))
print(f"S = {integr} = {integr.evalf():.2f}")
tv = np.linspace(0, 2*np.pi, 1000)
xv = 2*np.cos(tv)
yv = 3*np.sin(tv)
plt.plot(xv, yv)
plt.axis("equal")
plt.title("Б) 2*cos(t), 3*sin(t)")
plt.grid()
plt.show()
```

**Упражнение 4**



```python
fi = Symbol("f")
r = exp(fi)
integr = 1/2*integrate(r**2, (fi, 0, 2*pi))
print(f"S = {integr} = {integr.evalf():.2f}")

fi=np.linspace(0,2*np.pi,100)
r=np.exp(fi)
x=r*np.cos(fi)+1
y=r*np.sin(fi)
plt.plot(x,y)
plt.grid(True)
```

**Упражнение 5**



```python
x = Symbol('x')
y = x**2
a = 1
b = 2
dy_dx = diff(y, x, 1)
integr = integrate(sqrt(1+(dy_dx)**2), (x, a, b))
print(f"l = {integr} = {integr.evalf():.3f}")
```

**Упражнение 6**



```python
t = Symbol('t')
x = 2*cos(t)
y = 3*sin(t)
dx_dt = diff(x, t, 1)
dy_dt = diff(y, t, 1)
integr = integrate(sqrt(dx_dt**2 + dy_dt**2), (t, 0, 2*pi))
print(f"l = {integr} = {integr.evalf():.3f}")
```

```python
x = Symbol('x')

y1 = (x - 2)**2
y2 = 4 - x**2

a = 0
b = 2

Vx = pi * integrate(y2**2 - y1**2, (x, a, b))
print(f"Объем Ox: {Vx} = {Vx.evalf():.3f}")

Vy = 2 * pi * integrate(x*(y2 - y1), (x, a, b))
print(f"Объем Oy: {Vy} = {Vy.evalf():.3f}")


xv = np.linspace(0, 2, 1000)
f1 = (xv - 2)**2
f2 = 4 - xv**2

plt.plot(xv, f1)
plt.plot(xv, f2)
plt.fill_between(xv, f1, f2, alpha=0.3)

plt.grid(True)
plt.show()
```

**!!Упражнение 1С!!**



```python
x = Symbol('x')
y1 = x**2 + 1
y2 = 3 - x
y3 = 1
intersect1 = solve(Eq(y3, y1), x)[0]
intersect2 = solve(Eq(y1, y2), x)[1]
intersect3 = solve(Eq(y2, y3), x)[0]
integr1 = integrate(y1 - y3, (x, intersect1, intersect2))
integr2 = integrate(y2 - y3, (x, intersect2, intersect3))
print(f"S = {integr1 + integr2} = {(integr1 + integr2).evalf():.3f}")

x=np.linspace(-3,3,1000)
plt.plot(x, x**2+1)
plt.plot(x, 3-x)
plt.plot(x, 0*x+1)
plt.axis("equal")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Графики функций x**2+1, 3-x, 1")
plt.grid()
```

**Упражнение 2С**



```python
t = Symbol('t')
x = cos(t)**3
y = sin(t)**3
dx_dt = diff(x, t, 1)
integr = -integrate(y*dx_dt, (t, 0, pi))*2
print(f"S = {integr} = {integr.evalf():.3f}")

tv = np.linspace(0, 2*np.pi, 1000)
xv = np.cos(tv)**3
yv = np.sin(tv)**3
plt.plot(xv, yv)
plt.axis("equal")
plt.title("cos(t)**3, sin(t)**3")
plt.grid()
plt.axis("equal")
plt.xlabel("x=cos(t)**3")
plt.ylabel("y=sin(t)**3")
plt.show()
```
